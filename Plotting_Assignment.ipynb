```python
# Load the dataset
file_path = '/content/dataset/fuel_econ.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows to understand the structure of the data
print(data.head(10))
```

```python
# Set up plot style
plt.style.use('seaborn-darkgrid')
```

```python
# Plot histogram for fuel efficiency (combined)
plt.figure(figsize=(10, 6))
sns.histplot(data['comb'], bins=30, kde=True, color='blue')
plt.title('Distribution of Combined Fuel Efficiency (mpg)')
plt.xlabel('Combined Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histogram for CO2 emissions with various bin widths
plt.figure(figsize=(10, 6))
sns.histplot(data['co2'], bins=30, kde=True, color='red')
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (grams per mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['co2'], bins=50, kde=True, color='green')
plt.title('Distribution of CO2 Emissions (50 Bins)')
plt.xlabel('CO2 Emissions (grams per mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histogram for 'city' fuel efficiency
plt.figure(figsize=(10, 6))
sns.histplot(data['city'], bins=30, kde=True, color='blue')
plt.title('Distribution of City Fuel Efficiency (mpg)')
plt.xlabel('City Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Histogram for 'highway' fuel efficiency
plt.figure(figsize=(10, 6))
sns.histplot(data['highway'], bins=30, kde=True, color='purple')
plt.title('Distribution of Highway Fuel Efficiency (mpg)')
plt.xlabel('Highway Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Heatmap of correlations between numeric columns
plt.figure(figsize=(12, 8))
corr = data[['city', 'highway', 'comb', 'co2', 'feScore', 'ghgScore']].corr()
```

```python
# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlations Between Fuel Efficiency and Environmental Metrics')
plt.show()
```
